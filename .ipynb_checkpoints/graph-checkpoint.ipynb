```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results.csv")

df["cooldown"] = df["cooldown"].astype(float)
df["time"] = df["time"].astype(float)

avg_by_cooldown = (
    df.groupby("cooldown", as_index=True)["time"]
      .mean()
      .sort_index()
)

plt.figure(figsize=(8, 5))
avg_by_cooldown.plot(kind="bar")
plt.title("Average Time to Win by Blue Dodge Cooldown")
plt.xlabel("Blue dodge cooldown")
plt.ylabel("Average time to win (s)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
